```python
import matplotlib.pyplot as plt
import numpy as np
import os

PATH = os.getcwd()

year = 3.2e7 # Anos (1 year = 3.2x10^7 segundos)
AU = 1.5e11 # Unidade Astronomica (1 AU = 1.5x10^11 m)
```

```python
def vx(vxi, xi, dt, ri):
    return (vxi - (4 * np.pi**2 * xi ) / (ri**3) * dt)

def vy(vyi, yi, dt, ri):
    return (vyi - (4 * np.pi**2 * yi ) / (ri**3) * dt)
```

```python
N = 50000

xi = np.zeros(N)
yi = np.zeros(N)

xi[0] = 1.
yi[0] = 0.0

vxi = np.zeros(N)
vyi = np.zeros(N)

vxi[0] = 0
vyi[0] = 2 * np.pi / np.sqrt(xi[0]**2 + yi[0]**2) 

for i in range(0, len(vxi)-1):
    ri = np.sqrt(xi[i]**2 + yi[i]**2)

    vxi[i+1] = vx(vxi[i], xi[i], 0.002, ri)
    vyi[i+1] = vy(vyi[i], yi[i], 0.002, ri)

    xi[i+1] = xi[i] + vxi[i+1]*0.002
    yi[i+1] = yi[i] + vyi[i+1]*0.002

plt.grid()
plt.scatter(xi,yi)
plt.show()
```

```python
arq = open("teste_python.txt", "w")

for i in range(0, len(xi)):
    arq.write(f'{xi[i]:.7},{yi[i]:.7}\n')
    arq.flush()

arq.close()
```
