```python
import matplotlib.pyplot as plt
import numpy as np
import os

PATH = os.getcwd()

year = 3.2e7 # Anos (1 year = 3.2x10^7 segundos)
AU = 1.5e11 # Unidade Astronomica (1 AU = 1.5x10^11 m)
G = 6.6743e-11 # Constante de Gravidade
M_sun = 1.989e30  # mass do sol em kg
```

```python
def vx(vxi, xi, dt, ri):
    return (vxi - (4 * np.pi**2 * xi ) / (ri**3) * dt)

def vy(vyi, yi, dt, ri):
    return (vyi - (4 * np.pi**2 * yi ) / (ri**3) * dt)
```

```python
N = 50000

xi = np.zeros(N)
yi = np.zeros(N)

xi[0] = 1.
yi[0] = 0.0

vxi = np.zeros(N)
vyi = np.zeros(N)

vxi[0] = 0
vyi[0] = 2 * np.pi / np.sqrt(xi[0]**2 + yi[0]**2) 

for i in range(0, len(vxi)-1):
    ri = np.sqrt(xi[i]**2 + yi[i]**2)

    vxi[i+1] = vx(vxi[i], xi[i], 0.002, ri)
    vyi[i+1] = vy(vyi[i], yi[i], 0.002, ri)

    xi[i+1] = xi[i] + vxi[i+1]*0.002
    yi[i+1] = yi[i] + vyi[i+1]*0.002

plt.grid()
plt.scatter(xi,yi)
plt.show()
```

```python
N = 50000

xi = np.zeros((2,N))
yi = np.zeros((2,N))

xi[0][0] = 1.
yi[0][0] = 0.0

xi[1][0] = 1.52
yi[1][0] = 0.0

vxi = np.zeros((2,N))
vyi = np.zeros((2,N))

vxi[0][0] = 0
vyi[0][0] = 2 * np.pi / np.sqrt(xi[0][0]**2 + yi[0][0]**2)

vxi[1][0] = 0
vyi[1][0] = 2 * np.pi / np.sqrt(xi[1][0]**2 + yi[1][0]**2)

for i in range(0, 2):
    for j in range(0, N-1):
        ri = np.sqrt(xi[i][j]**2 + yi[i][j]**2)

        vxi[i][j+1] = vx(vxi[i][j], xi[i][j], 0.002, ri)
        vyi[i][j+1] = vy(vyi[i][j], yi[i][j], 0.002, ri)

        xi[i][j+1] = xi[i][j] + vxi[i][j+1]*0.002
        yi[i][j+1] = yi[i][j] + vyi[i][j+1]*0.002

plt.grid()
plt.scatter(xi[0],yi[0], color='blue')
plt.scatter(xi[1],yi[1], color='red')
plt.show()
```
